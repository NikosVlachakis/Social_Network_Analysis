```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python

random.seed(seedno)
# Each node in the ring is joined to its 4 nearest neighbors 
d = 4
# n is the number of nodes in the network
n = 41
# select suitable p value for the rewiring probability
p = 0
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx 
REG = nx.watts_strogatz_graph(n,d,p,seed=seedno)
#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG, with_labels=True)
# Show the plot
plt.show()
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)

# In the Gnm model, a graph is chosen uniformly at random from the set of all graphs with n nodes and M edges.
n = 41 
M = 150
# for the RGER use the gnm_random_graph from networkx
RGER = nx.gnm_random_graph(n,M,seed=seedno)
#visualize the graph
nx.draw_circular(RGER, with_labels=True)
# Show the plot
plt.show()
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n = 41
R = 0.35

# for the RGG use the random_geometric_graph from networkx
RGG = nx.random_geometric_graph(n, R,seed=seedno)
# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n = 41
d = 4
# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF = nx.barabasi_albert_graph(n,d,seed=seedno)
#visualize the graph 
nx.draw_circular(SF, with_labels=True)
# Show the plot
plt.show()
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n = 41
d = 4
p = 0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW = nx.watts_strogatz_graph(n,d,p,seed=seedno)
#visualize the graph
nx.draw_circular(SW, with_labels=True)
# Show the plot
plt.show()
```

## B. Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
# we create a list with objects of each topology
topologies = [(REG, "REG"), (RGER, "RGER"), (RGG, "RGG"), (SF, "SF"), (SW, "SW")]
```

```python
def cc(G,name): 
    # compute the clustering coefficient for each node and show the histogram
    cc = nx.clustering(G)
    # take the values of the dictionary and put them in a list
    cc_list = list(cc.values())
    # plot the histogram
    plt.hist(cc_list,bins=20)
    # set the title
    plt.title(name)
    # x-axis label
    plt.xlabel("Clustering coefficient")
    # y-axis label
    plt.ylabel("Nodes")
    # show the plot
    plt.show()

def cc_avg(G):
    # compute the average clustering coefficient
    cc_avg = nx.average_clustering(G)
    print("The average clustering coefficient is: ", cc_avg)
    # compute the variance of the clustering coefficient
    cc_var = np.var(list(nx.clustering(G).values()))
    print("The variance of the clustering coefficient is: ", cc_var)

def cc_min_max(G):
    # compute the clustering coefficient for each node
    cc = nx.clustering(G)
    # take the values of the dictionary and put them in a list
    cc_list = list(cc.values())
    # all indexes with minimum and maximum values
    min_indexes = [i for i, j in enumerate(cc_list) if j == min(cc_list)]
    max_indexes = [i for i, j in enumerate(cc_list) if j == max(cc_list)]
    # print the nodes with minimum and maximum values
    print("The nodes with minimum clustering coefficient are: ", min_indexes)
    print("The nodes with maximum clustering coefficient are: ", max_indexes)


# do the same for all the graphs
for G,name in topologies:
    cc(G,name)
    cc_avg(G)
    cc_min_max(G)


```

### Result analysis

- We can see that the the average and per node clustering coeffient for the REG topology is 0.5. The fact that the average and the per node clustering coefficient match is because each node has the same number of neighbors. The variance in the case of REG is 0 because all nodes have the same clustering coefficient, so the distance between each node from the average is 0. As about the RGER topology the low average clustering coefficient(ACC) is due to the fact that the number of edges is too small (150). If we take account that the number of nodes is 41, then for a fully connected graph we should have 1640 edges. Concerning the RGG topology, the threshold keeps the graph dense, so the ACC is high. The distribution for the SW topology tends to be uniform around the (0.2-0.3). Low values are expected for this kind of topology because it's main characteristic is the small number of steps needed to reach any node from any other node. In this kind of graph most nodes are not neighbors of one another, so the ACC is low. The scale free topology follows the distribution of the SW topology, so the ACC is low. Moving on the variance, as we can see from the above graphs, it keeps a low value, which is what we expected as the distribution for each topology tends to be uniform. 

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
def compute_path_lengths_mean_var(G):
    path_lengths = []
    for u,v in nx.all_pairs_shortest_path_length(G):
        path_lengths.extend(v.values())
    print("Mean =",np.mean(path_lengths),"Variance =", np.var(path_lengths))

def compute_diameter(G):
    print("Diameter =",nx.diameter(G))

def compute_radius(G):
    print("Radius =",nx.radius(G))

for G,name in topologies:
    print(name + ":")    
    compute_path_lengths_mean_var(G)
    compute_diameter(G)
    compute_radius(G)
```

### Result analysis

-  As we can see from the above results the average path length(APL) for the REG topology is the largest comparing to the other topologies. This is what we expected as the connection between different nodes is not the primary focus for the REG topology, comparing to other type of topologies, such as the SW. The RGG topology has the second longest APL after the regular graph topology. The other three topologies (SW,SF,RGER) keep the APL in low rates. The diameter and radius of RGER,RGG,SF,SW topologies have approximately the same values, which is logic as the APL for these topologies varies at the same rate. The diameter and radius of the REG topology are the largest, which is what we expected as the APL for this topology is the largest.

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G,name):
    diameter = nx.diameter(G)
    nodes = []
    for u,v in nx.all_pairs_shortest_path_length(G):
        for key, value in v.items():
            if value == diameter:
                nodes.append(key)
    node = max(set(nodes), key = nodes.count)
    nodes = []
    for u,v in nx.all_pairs_shortest_path_length(G):
        for key, value in v.items():
            if value == diameter and key == node:
                nodes.append(u)
    print(name + ":")
    print("Node with the most connections of minimum path length that is equal to the diameter of the graph is: ", node)
    print("The set of nodes that are connected with this node is: ", nodes)
    

# do the same for all the graphs
for G,name in topologies:    
    find_nodes(G,name)
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G,name):
    print(name + ":")
    print("The periphery is: ",nx.periphery(G))

for G,name in topologies:
    find_periphery(G,name)    
```

```python
def find_center(G,name):
    print(name + ":")
    print("The center is: ",nx.center(G))

for G,name in topologies:
    find_center(G,name)
```

### Result analysis

-  For REG topology is obvious that the center and the periphery are the same. Because of the costant topology, the minimum and maximum distances between a vertex to all other vertices has the same value. For the rest topologies because of the random nature of the graphs, there is not the same pattern between the center and the periphery.

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_degree(G,name):
    degree = nx.degree_centrality(G)
    # take the values of the dictionary and put them in a list
    degree_list = list(degree.values())
    # plot the histogram
    plt.hist(degree_list,bins=20)
    # set the title
    plt.title(name)
    # x-axis label
    plt.xlabel("Degree centrality")
    # y-axis label
    plt.ylabel("Number of nodes")
    # show the plot
    plt.show()

def find_degree_avg(G,name):
    degree = nx.degree_centrality(G)
    # take the values of the dictionary and put them in a list
    degree_list = list(degree.values())
    # compute the average degree centrality
    degree_avg = np.mean(degree_list) 
    print(name + ":")
    print("The average degree centrality is: ", degree_avg)
    # compute the variance of the degree centrality
    degree_var = np.var(degree_list)
    print("The variance of the degree centrality is: ", degree_var)

   


for G,name in topologies:
    find_degree(G,name)
    find_degree_avg(G,name)
    
```

### Result analysis

-  The fact that the variance of degree centality tends to be 0 for all different kind of topologies, shows that the degree of the nodes is almost the same for all nodes. As about the REG topology, it is clear that takes the lowest value of degree centality as the number of neighbors is fixed, comparing to the other topologies. 

### Γ2. Κεντρικότητα εγγύτητας (Closeness centrality)
##### Γ2.1 Να αναπαρασταθεί η κατατανομή της εγγύτητας κόμβων για κάθε τοπολογία.
##### Γ2.2 Να υπολογιστεί η μέση εγγύτητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_closeness(G,name):
    closeness = nx.closeness_centrality(G)
    # take the values of the dictionary and put them in a list
    closeness_list = list(closeness.values())
    # plot the histogram
    plt.hist(closeness_list,bins=20)
    # set the title
    plt.title(name)
    # x-axis label
    plt.xlabel("Closeness centrality")
    # y-axis label
    plt.ylabel("Number of nodes")
    # show the plot
    plt.show()

def find_closeness_avg(G,name):
    closeness = nx.closeness_centrality(G)
    # take the values of the dictionary and put them in a list
    closeness_list = list(closeness.values())
    # compute the average closeness centrality
    closeness_avg = np.mean(closeness_list)
    print(name + ":")
    print("The average closeness centrality is: ", closeness_avg)
    # compute the variance of the closeness centrality
    closeness_var = np.var(closeness_list)
    print("The variance of the closeness centrality is: ", closeness_var)

for G,name in topologies:
    find_closeness(G,name)
    find_closeness_avg(G,name)
    
```

### Result analysis

-  As we described in the previous markdown, the variance again is close to 0 for all topologies. This shows that the closeness centrality is almost the same for all nodes. The distribution for RGER and RGG seems to be uniform and the results are very close between them. As about the REG topology, the value of closeness centrality is the lowest, as all the nodes are part of the periphery of the graph. The SF topology gives the highest value of closeness centrality. Lastly, the value of SW's average closeness centrality is not that high as we expected, because most nodes are not neighbors of one another, so the distance between them is getting bigger, especially as the number of nodes increases.

### Γ3. Ενδιαμεσική κεντρικότητα (Betweenness centrality)
##### Γ3.1 Να αναπαρασταθεί η κατατανομή της ενδιαμεσικής κεντρικότητας κόμβων για κάθε τοπολογία.
##### Γ3.2 Να υπολογιστεί η μέση ενδιαμεσική κεντρικότητα καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_betweenness(G,name):
    betweenness = nx.betweenness_centrality(G)
    # take the values of the dictionary and put them in a list
    betweenness_list = list(betweenness.values())
    # round 2 decimals for better visualization
    betweenness_list = [round(num, 5) for num in betweenness_list]
    # show the histogram of the betweenness centrality
    plt.hist(betweenness_list,bins=20)
    # set the title
    plt.title(name)
    # x-axis label
    plt.xlabel("Betweenness centrality")
    # y-axis label
    plt.ylabel("Number of nodes")
    # show the plot
    plt.show()
    


def find_betweenness_avg(G,name):
    betweenness = nx.betweenness_centrality(G)
    # take the values of the dictionary and put them in a list
    betweenness_list = list(betweenness.values())
    # compute the average betweenness centrality
    betweenness_avg = np.mean(betweenness_list)
    print(name + ":")
    print("The average betweenness centrality is: ", betweenness_avg)
    # compute the variance of the betweenness centrality
    betweenness_var = np.var(betweenness_list)
    print("The variance of the betweenness centrality is: ", betweenness_var)

for G,name in topologies:
    find_betweenness(G,name)
    find_betweenness_avg(G,name)
```

### Result analysis

-  The average betweenness centrality is the highest for the REG topology, comparing to the other topologies. The fact that each vertex is connected with k neighbors, in our case 4 which is a low number, makes the betweenness centrality to be high. This is because the small number of neighbors makes a vertex to be a very possible bridge between the shortest path of two other vertices. As we can see to the other topologies which they have much smaller values, the randomness of the graph makes the betweenness centrality to be lower. Again the variances are very close to 0, which shows that the betweenness centrality is almost the same for all nodes.

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G,name):
    # Alpha value of the Katz centrality using the maximum eigenvalue of the adjacency matrix
    alpha = (1/max(nx.adjacency_spectrum(G))) - 0.01
    # compute the Katz centrality
    katz = nx.katz_centrality(G, alpha, beta=1.0)
    # take the values of the dictionary and put them in a list
    katz_list = list(katz.values())
    # plot the histogram
    plt.hist(katz_list,bins=20)
    # set the title
    plt.title(name)
    # x-axis label
    plt.xlabel("Katz centrality")
    # y-axis label
    plt.ylabel("Number of nodes")
    # show the plot
    plt.show()

def find_katz_avg(G,name):
    # Alpha value of the Katz centrality using the maximum eigenvalue of the adjacency matrix
    alpha = (1/max(nx.adjacency_spectrum(G))) - 0.01
    # compute the Katz centrality
    katz = nx.katz_centrality(G, alpha)
    # take the values of the dictionary and put them in a list
    katz_list = list(katz.values())
    # compute the average Katz centrality
    katz_avg = np.mean(katz_list)
    print(name + ":")
    print("The average Katz centrality is: ", katz_avg.real)
    # compute the variance of the Katz centrality
    katz_var = np.var(katz_list)
    print("The variance of the Katz centrality is: ", katz_var)

for G,name in topologies:
    find_katz(G,name)
    find_katz_avg(G,name)
```

### Result analysis

-  Katz centrality remains approximately the same for all topologies with variance close to 0. With this algorithm we achieve average centrality very similar for all topologies.  

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G,name):
    # compute the most central node depending on the degree centrality, closeness centrality, betweenness centrality and Katz centrality
    degree = nx.degree_centrality(G)
    closeness = nx.closeness_centrality(G)
    betweenness = nx.betweenness_centrality(G)
    # Alpha value of the Katz centrality using the maximum eigenvalue of the adjacency matrix
    alpha = (1/max(nx.adjacency_spectrum(G))) - 0.01
    katz = nx.katz_centrality(G, alpha)

    # find the max of degree , closeness, betweenness and katz
    degree_max = max(degree, key=degree.get)
    closeness_max = max(closeness, key=closeness.get)
    betweenness_max = max(betweenness, key=betweenness.get)
    katz_max = max(katz, key=katz.get)

    # print the results
    print(name + ":")
    print("The most central node according to the degree centrality is: ", degree_max)
    print("The most central node according to the closeness centrality is: ", closeness_max)
    print("The most central node according to the betweenness centrality is: ", betweenness_max)
    print("The most central node according to the Katz centrality is: ", katz_max)
    print("----------------------------------------------------------------------------")

   
for G,name in topologies:
    find_central(G,name)


```

### Result analysis

-  As we can see from the above results, for each topology the most central node is the same node in most cases. In SF and RGER the only node that differs from the others in the node computed by the betweeness centrality. This is because the algorithm is based on the shortest path between two nodes, so the node that is in the middle of the shortest path gets higher betweenness centrality. The REG topology because of the symmetry has the same central nodes regardless of the algorithm.
